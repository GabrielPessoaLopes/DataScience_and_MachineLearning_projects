```python
#Counts missing values in the sample
missing_data = df.isnull()
missing_data.head(5)
#Displays the missing values in "children" - False represents the not missing values
print (missing_data["adults"].value_counts())
```

```python
#"adults" has 4 missing values
df["adults"].replace('?', np.NaN, inplace = True) 
print(df["adults"])

#replaces missing values with the mean
avg_norm_loss = df["adults"].astype("float").mean(axis=0)
df["adults"].replace(np.nan, avg_norm_loss, inplace=True) 
print(df["adults"])

#replaces missing values with the nan
df["adults"].replace(avg_norm_loss, np.nan, inplace=True) 
df.reset_index(drop=True, inplace=True) 
print(df["adults"])

#removes the missing values
df.dropna(subset=["adults"], axis=0, inplace=True) 
df.reset_index(drop=True, inplace=True) 
print(df["adults"])
```

```python
#Displays sample's variables types and other info 
df.dtypes
df.info()
```

```python
df[["is_canceled", "is_repeated_guest"]] = df[["is_canceled", "is_repeated_guest"]].astype("bool")
df[["adults"]] = df[["adults"]].astype("int64")
df[["agent", "company"]] = df[["agent", "company"]].astype("object")
df[["reservation_status_date"]] = df[["reservation_status_date"]].astype("datetime64")
df.dtypes 
df.info()
```

```python
#normalizes a variable to vary from 0 to 1
df['days_in_waiting_list (0 and 1)'] = df['days_in_waiting_list']/df['days_in_waiting_list'].max()
df[['days_in_waiting_list (0 and 1)']].info()
```

```python
#Creates three sets of values from another variable
bins = np.linspace(min(df["days_in_waiting_list"]), max(df["days_in_waiting_list"]),4) 
group_names = ['Low', 'Medium','High'] 
df["days_in_waiting_list_classes"] = pd.cut(df["days_in_waiting_list"], bins, labels=group_names,include_lowest=True )
df[["days_in_waiting_list", "days_in_waiting_list_classes"]].head(5)
```

```python
#Creates a histogram from the created column
plt.hist(df["days_in_waiting_list_classes"])
plt.xlabel("Days in the waiting list by class")
plt.ylabel("Values")
plt.title("Days in the waiting list histogram") 
```

```python
#shows a pivot table between a scale variable and two categorical variables
df_new = df[["adults","hotel_type","arrival_date_month"]] 
grouped = df_new.groupby(["adults","hotel_type","arrival_date_month"],as_index=False).count()
grouped_pivot = grouped.pivot(index="adults",columns=["hotel_type","arrival_date_month"])
print(grouped_pivot)
```

```python
#Creates heatmap from the previous pivot table
plt.pcolor(grouped_pivot, cmap='RdBu')
plt.colorbar()
plt.show()
```

```python
#Displays the scale variable's mean for each categorical variable
df_groups = df[["adults","hotel_type","arrival_date_month"]]
grouped_by_mean = df_groups.groupby(["adults","hotel_type","arrival_date_month"],as_index=False).mean()
print(grouped_by_mean)
```

```python
# Performs the ANOVA test
import scipy.stats as stats

# Group by "arrival_date_month" and get the groups for "adults"
grouped = df.groupby(["arrival_date_month"])["adults"]

# Collect the groups in a list
groups = [grouped.get_group(month) for month in df["arrival_date_month"].unique()]

# Perform ANOVA
f_val, p_val = stats.f_oneway(*groups)

# Print the results
print(f"ANOVA F-value: {f_val}")
print(f"P-value: {p_val}")
```
